```python
import pandas as pd
from sklearn import tree

df = pd.read_csv("titanic.csv")
X = df.drop(["Name", "SibSp","Parch","Ticket","Cabin","Embarked","PassengerId","Pclass","Survived"], axis = "columns")
Y = df["Survived"]

```

Encoding "Sex" using Pandas

```python
Sex_1 = pd.get_dummies(X["Sex"])
Sex_1
X_new = pd.concat([Sex_1, X], axis="columns")
X_new
X_1 = X_new.drop(["Sex"], axis = "columns")
X_1
X_1 = X_new.fillna(X_new["Age"].mean())
X_1.head(10)
```

Encoding "Sex" using OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
X_ES = encoder.fit_transform(X[["Sex"]])
encoded_sex_df = pd.DataFrame(X_ES.toarray(),columns = encoder.get_feature_names_out(["Sex"]))
encoded_sex_df
X_2 = pd.concat([encoded_sex_df, X], axis = "columns")
X_2 = X_2.drop(["Sex"], axis = "columns")
X_2
```

Filling up the missing ages

```python
X_2["Age"].fillna(X_2["Age"].mean(), inplace=True)
X_2.head(10)
```
